# The Experiment
## Abstract
* Clinical symptoms such as fever indicate exposure to pathogens.
    * During "infectious prodomal period" - occurs after incubation period, this is when pathogens multiply
    * Machine learning method from this paper detects asymptomatic states during incubation, which is when symptoms first begin to develop
* The data: physiological waveform data from primate studies of
    * Viral exposure: Ebola, Marburg, Lassa, Nipah viruses
    * Bacterial exposure: Y. pestis (plague)
* In most subjects, detection was achieved well before fever -> 51h early warning time on average
    * Which physiological features are most informative for early dection?

## Introduction
* Assess pathogen exposure based on host physiological waveforms, not diagnostics based on fever or biomolecules of the pathogen itself
    * Advantages: earlier patient care increases probability of a positive prognosis (projection) -> measures to reduce transmission of disease
* Following pathogen exposure: incubation phase can vary, 3-25 days for hermorrhagic (escape of blood from a ruptured blood vessel) fevers, 2-4 days for Y. pestis
* Goal: build an "ideal" sensor system capable of detecting pathogen exposure during earliest moments of incubation period
    * Hypothesis: quantifiable abnormalities in physiological waveforms (electrocardiography, hemodynamics, temperature) before clinical signs could be basis for ideal signals
    * Diagnosis is usually based on identification of pathogen-specific molecular signatures (RNA or DNA sequencing, culture, which is the propagation of microorganisms, etc.) in a fluid.
        * Genomic and transcriptional profiles have been promising for results, but have logistic hurdles
* These results, in non-clinical contexts, could lead to wearable sensor and monitoring systems
    * More processing power is available to consumers to make use of the data
    * These algorithms can enable this vision

## What is the data being used?
* High-resolution (fast sampling rates and quantized amplitudes) physiological data from non-human primates (NHPs) exposured via intramuscular (IM), aerosol, or intratracheal (see: trachea) routes to one of several viral fevers:

        * Ebola virus (EBOV)
        * Marburg virus (MARV)
        * Lassa virus (LASV)
        * Nipah virus (NiV)

* One bacterial pathogen:
        
        * Y. pestits
* Build high-sensitivity, low specificty (not informative of particular pathogens) processing and detection algorithm -- in given paper only
* Data is standardized to remove rhythms, aggregated (collected into summary reports) to not take into account short-term fluctuations
* Telemetry devices were implanted 3 to 5 months before exposure
* Subjects exposued under sedation (administration of a drug) via aerosol, intramuscular injection, or intratracheal exposure, depending on the study

### Data processing
* Remove artifacts, divide sequential time series into minute intervals, ommitting top and bottom 2% quantiles for each
* Estimate baseline diurnal (daily) statistics during the pre-exposure period -> assumed consistency

## Plotting Feature Time Series for Each Subject
Do the trends for a certain feature vary between the study subjects? Do the trends change when the exposure period starts?

```python
def plot_feature(dataset, feature):
    subject_list = np.unique(dataset['SubjectName'])

    for subject in subject_list:
        plt.plot('TimeS', feature, 
        data=dataset[dataset['SubjectName'] == subject],
        lw=0.8)
        line_x = dataset[(dataset['SubjectName'] == subject) &
        dataset['Y'] == 1]['TimeS'].to_list()[0]
        # print("Exposure period starts for " + subject + " at: " + str(line_x) + " time steps")
        plt.plot((line_x, line_x), (min(dataset[feature]), max(dataset[feature])), alpha=0.7, ls='--', lw=2)

    plt.legend(subject_list)
    plt.title(feature + " for Each Subject", fontsize=20)
    plt.xlabel("Time Steps", fontsize=14)
    plt.xticks(rotation='vertical')
    plt.ylabel(feature, fontsize=14)

plot_feature(data, "Temp_Mean")
```

```python
plot_feature(data, "RespMean_Q25")
```

```python
plot_feature(data, "RespMean_Q75")
```

# Run Analysis on Rest of the Datasets

```python
def analysis(dataset):
    print('Analysis for ' + dataset['Study'][0] + '\n')
    print('Shape of data: \n')
    print(str(dataset.shape) + '\n')
    print('Dtypes:\n')
    print(str(dataset.get_dtype_counts()) + '\n')
    print('Number of Days of Data Collection:\n')
    num_days = len(np.unique(dataset['Day']))
    print(str(num_days) + '\n')
    print('Subjects:\n')
    print(np.unique(dataset['SubjectName']))
    print('\n')
    print('Number of Data Points for Each Subject: \n')
    print(str(pd.value_counts(dataset['SubjectName'])))
    print('\n')
    print('TS Plot for First Physiological Feature: \n')
    plot_feature(dataset, dataset.columns[5])

analysis(pd.read_csv('EBOV-Cyno-Aero-2.csv-Table 1.csv'))
```
